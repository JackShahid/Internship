3. Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. Name, IMDB rating, Year 
of release).


```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
page = requests.get("https://www.imdb.com/list/ls047058982")
page
```

```python
Soup = bs(page.content)
Soup
```

```python
name = Soup.find_all("h3", "lister-item-header")
name
```

```python
len(name)
```

```python
Name = []

for i in name:
    for j in i.find_all("a"):
        
        Name.append(j.text.replace("\n", " "))
Name
```

```python
rating = Soup.find_all("div", "ipl-rating-star small")
rating
```

```python
len(rating)
```

```python
Rating = []

for i in rating:
    Rating.append(i.text.replace("\n",""))
    
Rating
```

```python
year = Soup.find_all("span","lister-item-year text-muted unbold")
year
```

```python
len(year)
```

```python
Year = []

for i in year:
    Year.append(i.text)
Year
```

```python
import pandas as pd
```

```python
movies = pd.DataFrame({})
movies["Movie_Name"]= Name
movies["Rating"]= Rating
movies["Year"]= Year
movies
```
