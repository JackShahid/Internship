1. Write a python program to display all the header tags from‘en.wikipedia.org/wiki/Main_Page’.

```python
from bs4 import BeautifulSoup
import requests
```

```python
Page = requests.get("https://en.wikipedia.org/wiki/Main_Page")
Page
```

```python
Soup = BeautifulSoup(Page.content)
Soup
```

```python
h1 = Soup.find_all("h1", class_ = "firstHeading")
h1
```

```python
h2 = Soup.find_all("h2", class_ = "mp-h2")
h2
```

```python
h3 = Soup.find_all("h3", class_ = "vector-menu-heading")
h3
```

```python
titles = Soup.find_all(["h1","h2","h3"])
print("List all header tags", *titles, sep = '\n\n')
```

```python
import pandas as pd
```

```python
wikipedia = pd.DataFrame({})

wikipedia['titles'] = titles

wikipedia
```
