9. Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title, 
company name, CTC, and apply date.


```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get("https://internshala.com/fresher-jobs")
page
```

```python
Soup = BeautifulSoup(page.content)
Soup
```

```python
p1 = Soup.find_all("div", "heading_4_5 profile")
p1
```

```python
profile = []
for i in p1:
    profile.append(i.text.replace("\n", " "))
profile
```

```python
c1 = Soup.find_all("div", "heading_6 company_name")
c1
```

```python
company = []
for i in c1:
    company.append(i.text.replace("\n", ""))
company
```

```python
s1 = Soup.find_all("div", "other_detail_item")
s1
```

```python
ctc = s1[1::3]
ctc
```

```python
CTC = []
for i in ctc:
    CTC.append(i.text.replace("\n", ""))
CTC
```

```python
date = Soup.find_all("div", "item_body")
date
```

```python
d1 = date[2::3]
d1
```

```python
apply= []
for i in d1:
    apply.append(i.text.replace("\n",""))
apply
```

```python
fresher = pd.DataFrame({})
fresher["Job_Title"] = profile[:40]
fresher["Company_Name"] = company[:40]
fresher["CTC"] = CTC[:40]
fresher["Apply_Date"]= apply[:40]
```
