4. Write a python program to scrap book name, author name, genre and book review of any 5 books from
‘www.bookpage.com’


```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
page = requests.get("https://bookpage.com/")
page
```

```python
Soup = bs(page.content)
Soup
```

```python
book_name = Soup.find_all("h4", class_="italic")
book_name
```

```python
Book_name = []
for i in book_name:
    Book_name.append(i.text.replace("\n", ""))
Book_name
```

```python
book_author = Soup.find_all("p", class_= "sans bold")
book_author
```

```python
Book_author = []
for i in book_author:
    Book_author.append(i.text.replace("\n", " "))
Book_author
```

```python
book_genre = Soup.find_all("p", "genre-links hidden-phone")
book_genre
```

```python
len(book_genre)
```

```python
Book_genre = []
for i in book_genre:
    Book_genre.append(i.text.replace("\n"," "))
Book_genre
```

```python
review = Soup.find_all("p", "excerpt")
review
```

```python
Review  = []
for i in review:
    Review.append(i.text.replace("\n"," "))
Review
```
