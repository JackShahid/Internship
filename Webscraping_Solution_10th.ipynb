```python
from bs4 import BeautifulSoup as bs
import requests as r
import pandas as pd
```

```python
page = r.get("https://www.nobroker.in/property/sale/delhi/Hauz%20Khas?searchParam=W3sibGF0IjoyOC41NDc4ODk3LCJsb24iOjc3LjIwMzEyNDcsInBsYWNlSWQiOiJDaElKSTgtaGZSTGlERGtSVFJOcHhpRzhudGciLCJwbGFjZU5hbWUiOiJIYXV6IEtoYXMifV0=&radius=2.0")
page
```

```python
Soup = bs(page.content)
Soup
```

```python
house = Soup.find_all("h2", "heading-6 font-semi-bold nb__1AShY")
house
```

```python
h1 = []
for i in house:
    h1.append(i.text)
h1
```

```python
len(h1)
```

```python
location = Soup.find_all("div", "nb__2CMjv")
location
```

```python
l1= []
for i in location:
    l1.append(i.text)
l1
```

```python
len(l1)
```

```python
area = Soup.find_all("div", "nb__3oNyC")
area
```

```python
a1 = []
for i in area:
    a1.append(i.text)
a1
```

```python
len(area)
```

```python
emi = Soup.find_all("div", "font-semi-bold heading-6")
emi
```

```python
e1 = []
for i in emi:
    e1.append(i.text)
e1
```
